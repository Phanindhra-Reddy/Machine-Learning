```python
import easyocr
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
img_path = 'blog_banner.jpg'
```

```python
reader = easyocr.Reader(['en'],gpu=False)
result = reader.readtext(img_path)
result
```

```python
top_left = tuple(result[0][0][0])
btm_right = tuple(result[0][0][2])
text = result[0][1]
font = cv2.FONT_HERSHEY_SIMPLEX
```

```python
img = cv2.imread(img_path)
img = cv2.rectangle(img,top_left,btm_right,(0,255,0),3)
img = cv2.putText(img,text,top_left,font,0.7,(0,0,0),2,cv2.LINE_AA)
plt.imshow(img)
plt.show()
```

```python
img = cv2.imread(img_path)

for detection in result:
  top_left = tuple(detection[0][0])
  btm_right = tuple(detection[0][2])
  text = detection[1]
  img = cv2.rectangle(img,top_left,btm_right,(0,255,0),4)
  img = cv2.putText(img,text,top_left,font,1.0,(0,0,0),4,cv2.LINE_AA)


plt.imshow(img)
plt.show()
```
